```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
#APPLE
appl = yf.Ticker('AAPL')
appl_data = appl.history(period="2y") #2個月:"2mo", 2天:"2d", 2年:"2y" 
appl_data
```

```python
#APPLE
appl = yf.Ticker('AAPL')
appl_data = appl.history(start = "2020-05-01", end = "2022-05-01")  
appl_data
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("AAPL")
plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(appl_data.index, appl_data["Close"], c='r')
plt.grid(axis="y")
plt.show()
```

```python
import plotly.express as pe

pe.line(appl_data, x = appl_data.index, y="Close", title="AAPL" )
```

```python
import requests as r
import pandas as pd
import random,time
from datetime import datetime, date
from io import StringIO
```

```python
url = "https://www.twse.com.tw/en/indicesReport/MI_5MINS_HIST?response=csv&date=20221012"
```

```python
r.get(url)
```

```python
res = r.get(url)
```

```python
res.text
```

```python
df = pd.read_csv(StringIO(res.text.replace("=", "")), 
      header=["Date" in l for l in res.text.split("\n")].index(True))
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
month_list = pd.date_range('2022-08-01','2022-10-01', freq='MS').strftime("%Y%m%d").tolist()
month_list
```

```python
df = pd.DataFrame()
df
```

```python
for month in month_list:
  url = "https://www.twse.com.tw/en/indicesReport/MI_5MINS_HIST?response=csv&date=" + month
  res = r.get(url)
  stock_df = pd.read_csv(StringIO(res.text.replace("=", "")), 
      header=["Date" in l for l in res.text.split("\n")].index(True))
  df = df.append(stock_df, ignore_index = True)
  time.sleep(random.randint(1,5))

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```
